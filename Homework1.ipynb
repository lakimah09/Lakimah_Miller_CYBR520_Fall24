```python
import pandas as pd

# Assuming 'df' is your DataFrame
# For example purposes, let's create a sample DataFrame
df = pd.DataFrame({
    'capitalAve': [200, 150, 400, 500, 250],
    'capitalTotal': [1000, 800, 2000, 2500, 1200],
    'num857': [3, 2, 6, 7, 5],
    'money': [100, 200, 300, 400, 250]
})

# Compute the correlation matrix
correlation_matrix = df.corr()

# To avoid looking at the diagonal (self-correlation), set the diagonal to NaN
import numpy as np
np.fill_diagonal(correlation_matrix.values, np.nan)

# Find the pair with the highest positive correlation and highest negative correlation
# For highest positive correlation
max_corr = correlation_matrix.max().max()  # Max value of the entire matrix
max_corr_pair = correlation_matrix.stack().idxmax()  # Pair corresponding to the max value

# For highest negative correlation
min_corr = correlation_matrix.min().min()  # Min value of the entire matrix
min_corr_pair = correlation_matrix.stack().idxmin()  # Pair corresponding to the min value

print(f"The two features with the highest positive correlation: {max_corr_pair} with a correlation of {max_corr}")
print(f"The two features with the highest negative correlation: {min_corr_pair} with a correlation of {min_corr}")
```

```python
import pandas as pd
from scipy.stats import pearsonr

# Assuming 'df' is your DataFrame, and 'remove' and 'internet' are numeric columns
# Example data
df = pd.DataFrame({
    'remove': [1, 0, 1, 1, 0],
    'internet': [1, 0, 1, 1, 0]
})

# Drop rows with NaN values (optional if there are missing values)
df = df.dropna(subset=['remove', 'internet'])

# Calculate the Pearson correlation coefficient between 'remove' and 'internet'
corr_coefficient, p_value = pearsonr(df['remove'], df['internet'])

# Display the result
print(f"Correlation coefficient between 'remove' and 'internet': {corr_coefficient}")
print(f"P-value: {p_value}")
```

```python
import pandas as pd
from scipy.stats import pearsonr

# Assuming 'df' is your DataFrame and contains 'num3d' and 'num000' columns
# Example data:
df = pd.DataFrame({
    'num3d': [1, 2, 3, 4, 5],
    'num000': [5, 4, 3, 2, 1]
})

# Check if the columns 'num3d' and 'num000' exist and are numeric
print(df[['num3d', 'num000']].dtypes)

# If they contain non-numeric values, convert them to numeric (if necessary)
df['num3d'] = pd.to_numeric(df['num3d'], errors='coerce')
df['num000'] = pd.to_numeric(df['num000'], errors='coerce')

# Drop rows with NaN values in either 'num3d' or 'num000'
df = df.dropna(subset=['num3d', 'num000'])

# Calculate the Pearson correlation coefficient
corr_coefficient, p_value = pearsonr(df['num3d'], df['num000'])

# Display the result
print(f"Correlation coefficient between 'num3d' and 'num000': {corr_coefficient}")
print(f"P-value: {p_value}")
```
